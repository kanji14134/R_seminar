```r
y.3<-unlist(y.1)
y.3
```

```r
y.3*2
```

```r
m<-seq(1:6)
plot(m,y.3*2)
```

```r
c<-cbind(m,y.3*2)
class(c)
names(c)
c
```

```r
c.df<-data.frame(c)
class(c.df)
names(c.df)<-c("x","y")
c.df
```

```r
c.df<-data.frame(c.df,log(c.df$y))
names(c.df)[3]<-"logy"
plot(c.df$logy,c.df$y)
```

```r
result<-lm(y~logy,data=c.df)
summary(result)
```

```r
plot(c.df$logy,c.df$y)
abline(result)
```
